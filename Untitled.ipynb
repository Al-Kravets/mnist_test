```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
numpy.random.seed(42)
```

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train.reshape(60000, 784)
X_train = X_train.astype('float32')
X_train /=255
y_train = np_utils.to_categorical(y_train, 10)
X_test = X_test.reshape(10000, 784)
X_test = X_test.astype('float32')
X_test /=255
y_test = np_utils.to_categorical(y_test, 10)
```

```python
model = Sequential()
model.add(Dense(784,activation= "relu", input_dim=784, kernel_initializer= "normal" ))
model.add(Dense(10, activation = "softmax", kernel_initializer= "normal"))
model.compile(loss = "categorical_crossentropy", optimizer = "SGD", metrics = ["accuracy"])
print(model.summary())
```

```python
model.fit(X_train, y_train, batch_size = 200, epochs = 100, verbose =1)
```

```python
predictions = model.predict(X_train)
predictions = np_utils.to_categorical(predictions)
```

```python
print(predictions)
```

```python

scores = model.evaluate(X_test, y_test, verbose = 0)
print("Точность %.2f%%" % (scores[1]*100))
```
